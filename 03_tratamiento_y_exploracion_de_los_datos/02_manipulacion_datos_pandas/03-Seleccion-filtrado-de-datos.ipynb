## Filtrado con operadores logicos

Trabajos que NO sean remotos



```python
df[df['onsite_remote'] == 'remote']
```

```python
#Muestra todos los onside_remote que no son remotos
df[~(df['onsite_remote'] == 'remote')]
```

## O sean de Amazon NI remotos

```python
df[df['company'] == 'Amazon']
```

```python
df[((df['company']=='Amazon')&(df['onsite_remote']=='remote'))]
```

```python
# Trabajos que no sean de amazon ni remotos
df[~((df['title'] == 'Amazon') & (df['onsite_remote'] == 'remote'))]
```
